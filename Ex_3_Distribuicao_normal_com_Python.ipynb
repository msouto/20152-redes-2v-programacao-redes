```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf

def edCF(x, mu, sigma):
  #score Z
  z = (x-mu) / (sigma * np.sqrt(2))
  y=0.5 * (1 + erf(z) )
  return y

#parametros da distribuição normal
mu = 0
sigma = 1

#Gerar os valores de x para calcular a edCF
x_values = np.linspace(-3,3,1000)

#Calcular a edCF para os valores de x
y_values = edCF(x_values,mu,sigma)


#plotar a edCF
plt.plot(x_values,y_values,label='edCF da Distribuição Normal')
plt.xlabel('Valores de X')
plt.ylabel('edcf(x)')
plt.title('Função de distribuição Cumultativa empírica (edCF) - Distribuição Normal')
plt.legend()
plt.grid(True)
plt.show()
```
